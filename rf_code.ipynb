```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
from pprint import pprint as print
```

```python
#load csv's
weather_df = pd.read_csv('merged_df.csv', index_col=[0])
```

```python
pd.set_option('display.max_rows', None)
weather_df.head(999)
```

```python
x_values = weather_df["Year"]
y_values = weather_df["USA Temperature (C)"]

plt.scatter(x_values,y_values)
plt.title('Avg Temp per Year in USA')
plt.xlabel=('Years')
plt.ylabel('USA Temperature (C)')
plt.minorticks_on()
plt.grid(which='major', linestyle='-')
plt.grid(which='minor', linestyle=':')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

```python
def linear_reg_plot(x_values,y_values):
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
    regress_values = x_values * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    plt.scatter(x_values,y_values)
    plt.plot(x_values,regress_values,"r-")
    plt.annotate(line_eq,(x_values.median(),y_values.median()),fontsize=15,color="red")
    print(f"The r-squared is: {rvalue}")
    plt.show()
    
usa = weather_df["USA Temperature (C)"]
aus = weather_df["AUS Temperature (C)"]
bra = weather_df["BRA Temperature (C)"]
```

```python
x_values = weather_df["Year"]
y_values = usa
plt.title("USA Temperature (C) over the Decades")
plt.ylabel("Temperature (C)")
print(linear_reg_plot(x_values,y_values))
```

```python
x_values = weather_df["Year"]
y_values = aus
plt.title("Australia Temperature (C) over the Decades")
plt.ylabel("Temperature (C)")
print(linear_reg_plot(x_values,y_values))
```

```python
x_values = weather_df["Year"]
y_values = bra
plt.title("Brazil Temperature (C) over the Decades")
plt.ylabel("Temperature (C)")
print(linear_reg_plot(x_values,y_values))
```
