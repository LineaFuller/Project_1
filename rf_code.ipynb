```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
from pprint import pprint as print
```

```python
#load csv's
usa_df = pd.read_csv('USA_df.csv')
aus_df = pd.read_csv('AUS_df.csv')
```

```python
usa_df.head()
```

```python
aus_df.head()
```

```python
x_values = usa_df["Month"]
y_values = usa_df["Average Temp (1980 - 1999)"]

plt.scatter(x_values,y_values)
plt.title('Avg Temperature of the Month between 1980 to 1999 in USA')
plt.xlabel('Month')
plt.ylabel('Temperature (C)')
plt.ylim(-15, 25)
plt.minorticks_on()
plt.grid(which='major', linestyle='-')
plt.grid(which='minor', linestyle=':')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

```python
x_values = aus_df["AUS Month"]
y_values = aus_df["Average Temp (1980 - 1999)"]

plt.scatter(x_values,y_values)
plt.title('Avg Temperature of the Month between 1980 to 1999 in AUS')
plt.xlabel('Month')
plt.ylabel('Temperature (C)')
plt.ylim(0, 50)
plt.minorticks_on()
plt.grid(which='major', linestyle='-')
plt.grid(which='minor', linestyle=':')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```
