```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy.stats import linregress
import numpy as np
import requests
from scipy.stats import linregress
import csv
```

```python
Data=pd.read_csv('merged_df.csv')

year=Data['Year']
USA_temp=Data["USA Temperature (C)"]
USA_Pre=Data["USA Precipitation (mm)"]
plt.plot(year,USA_temp, "-r", label="USA_temperature")
plt.plot(year,USA_Pre,"-b", label="USA_precipitation")
plt.xlabel("Year ")
plt.ylabel("USA mean Temperature")
plt.legend()
plt.title("USA Temperature & USA Precipitation")
plt.xticks(rotation = 45)
plt.savefig('USAtemp_pre.png')
plt.show()


```

```python
plt.scatter(USA_temp,USA_Pre)
plt.xlabel("USA Temp")
plt.ylabel("USA_precipitation(mm)")
plt.title("USA Temp vs USA Precipitation")
(slope, intercept, rvalue, pvalue, stderr) = linregress(USA_temp, USA_Pre)
regress_values =  USA_temp* slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(USA_temp,regress_values,"r-")
plt.annotate(line_eq,(6,58),fontsize=15,color="red")
print(f"The r-squared is: {rvalue**2}")

plt.show()
```
