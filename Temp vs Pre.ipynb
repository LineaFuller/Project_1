```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy.stats import linregress
```

```python
pd = pd.read_csv("merged_df.csv")
```

```python
USA_temp=pd["USA Temperature (C)"]
USA_Pre=pd["USA Precipitation (mm)"]

plt.scatter(USA_temp,USA_Pre)
plt.xlabel("USA Temp")
plt.ylabel("USA_precipitation(mm)")
plt.title("USA Temp vs USA Precipitation")
(slope, intercept, rvalue, pvalue, stderr) = linregress(USA_temp, USA_Pre)
regress_values =  USA_temp* slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(USA_temp,regress_values,"r-")
plt.annotate(line_eq,(6,58),fontsize=15,color="red")
print(f"The r-squared is: {rvalue**2}")
plt.show()
```
