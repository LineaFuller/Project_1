```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy.stats import linregress
import numpy as np
import requests
from scipy.stats import linregress
import csv
```

```python
Data=pd.read_csv('merged_df.csv')


bins = [1900, 1911, 1921, 1931, 1941, 1951, 1961, 1971,1981,1991,2012]
group_labels = ["1901-1910", "1911-1920", "1931-1930", "1931-1940", "1941-1050", "1951-1960", "1961-1970", "1971-1980","1981-1990", "1991-2012"]

Data["Year group"]=pd.cut(Data["Year"],bins,labels=group_labels)

Data= Data.drop('Unnamed: 0', 1)

mean_grouped_Data=Data.groupby(['Year group']).mean()
max_grouped_Data=Data.groupby(['Year group']).max()
max_grouped_Data
USA_mean_temp=mean_grouped_Data.iloc[:,1]
USA_max_temp=max_grouped_Data.iloc[:,1]
USA_mean_pre=mean_grouped_Data.iloc[:,5]
USA_max_pre=max_grouped_Data.iloc[:,5]
# USA_max_temp
# USA_mean_temp
# AUS_mean_temp=
# year_grouped_Data=Data.groupby(["Year group"])

# print(year_grouped_Data)
# counts=Data(['Year group']).size()
# counts
# mean=Data.groupby(['year group']).mean()
# Max_temp=Data.groupby(['year group']).max()
summary_df=pd.DataFrame({"USA_Mean_temp":USA_mean_temp,
                        "USA_Max_temp":USA_max_temp,
                         "USA_Mean_precipitation":USA_mean_pre,
                      "USA_Max_precipitation":USA_max_pre
                  
                        })
summary_df




```

```python

plt.plot(group_labels,USA_mean_temp)

plt.xlabel("Year Group")
plt.ylabel("USA mean Temperature")

plt.title("USA average Temperature")
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.plot(group_labels,USA_max_temp)
plt.xlabel("Year Group")
plt.ylabel("USA Max Temperature")
plt.title("USA Max Temperature")
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.plot(group_labels,USA_mean_pre)

plt.xlabel("Year Group")
plt.ylabel("USA mean Precipitation")

plt.title("USA average Precipitation")
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.plot(group_labels,USA_max_pre)

plt.xlabel("Year Group")
plt.ylabel("USA Max Precipitation")

plt.title("USA Max Precipitation")
plt.xticks(rotation = 45)
plt.show()

```

```python


plt.scatter(USA_mean_temp,USA_mean_pre)
plt.xlabel("USA_mean_temp")
plt.ylabel("USA_mean_pre")
plt.title("USA_mean_temp vs USA_mean_pre")
(slope, intercept, rvalue, pvalue, stderr) = linregress(USA_mean_temp, USA_mean_pre)
regress_values =  USA_mean_temp* slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(USA_mean_temp,regress_values,"r-")
plt.annotate(line_eq,(6.9,42),fontsize=15,color="red")
print(f"The r-squared is: {rvalue**2}")
plt.show()
```

```python

plt.scatter(USA_max_temp,USA_max_pre)
plt.xlabel("USA_max_temp")
plt.ylabel("USA_max_pre")
plt.title("USA_max_temp vs USA_max_pre")
(slope, intercept, rvalue, pvalue, stderr) = linregress(USA_max_temp, USA_max_pre)
regress_values =  USA_max_temp* slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(USA_max_temp,regress_values,"r-")
plt.annotate(line_eq,(6.9,42),fontsize=15,color="red")
print(f"The r-squared is: {rvalue**2}")
plt.show()
```
