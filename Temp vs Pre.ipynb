```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy.stats import linregress
import numpy as np
import requests
from scipy.stats import linregress
import csv
```

```python
Data=pd.read_csv('merged_df.csv')
Data
year=Data['Year']
USA_temp=Data["USA Temperature (C)"]
USA_Pre=Data["USA Precipitation (mm)"]
plt.plot(year,USA_temp, "-r", label="USA_temperature")
plt.plot(year,USA_Pre,"-b", label="USA_precipitation")
plt.xlabel("Year ")
plt.ylabel("USA mean Temperature")
plt.legend()
plt.title("USA Temperature & USA Precipitation")
plt.xticks(rotation = 45)
plt.show()



```

```python
plt.scatter(USA_temp,USA_Pre)
plt.xlabel("USA Temp")
plt.ylabel("USA_precipitation(mm)")
plt.title("USA Temp vs USA Precipitation")
(slope, intercept, rvalue, pvalue, stderr) = linregress(USA_temp, USA_Pre)
regress_values =  USA_temp* slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(USA_temp,regress_values,"r-")
plt.annotate(line_eq,(6,58),fontsize=15,color="red")
print(f"The r-squared is: {rvalue**2}")
plt.show()
```

```python



bins = [1900, 1911, 1921, 1931, 1941, 1951, 1961, 1971,1981,1991,2012]
group_labels = ["1901-1910", "1911-1920", "1931-1930", "1931-1940", "1941-1050", "1951-1960", "1961-1970", "1971-1980","1981-1990", "1991-2012"]

Data["Year group"]=pd.cut(Data["Year"],bins,labels=group_labels)
Data



mean_grouped_Data=Data.groupby(['Year group']).mean()
max_grouped_Data=Data.groupby(['Year group']).max()

USA_mean_temp=mean_grouped_Data.iloc[:,1]
USA_max_temp=max_grouped_Data.iloc[:,1]
USA_mean_pre=mean_grouped_Data.iloc[:,5]
USA_max_pre=max_grouped_Data.iloc[:,5]

summary_df=pd.DataFrame({"USA_Mean_temp":USA_mean_temp,
                        "USA_Max_temp":USA_max_temp,
                         "USA_Mean_precipitation":USA_mean_pre,
                      "USA_Max_precipitation":USA_max_pre
                  
                        })
summary_df




```
