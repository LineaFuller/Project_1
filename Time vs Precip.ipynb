```python
# Import Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
# Read Precipitation csv file
pr_df = pd.read_csv("full_pr.csv", index_col=0)
pr_df
```

```python
# Analysis of USA Precipitation vs Time
pr_df.plot("Year", "USA Precipitation (mm)", kind="line", rot=45, title="Average Yearly Precipitation")
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, sharex=True)

ax1.plot(pr_df["Year"], pr_df["USA Precipitation (mm)"])
ax1.set_title("USA Precipitation (mm) vs Year")

ax2.plot(pr_df["Year"], pr_df["AUS Precipitation (mm)"])
ax2.set_title("AUS Precipitation (mm) vs Year")

ax3.plot(pr_df["Year"], pr_df["Brazil Precipitation (mm)"])
ax3.set_title("Brazil Precipitation (mm) vs Year")

ax2.tick_params(labelrotation=45)

plt.show()
```

```python
plt.plot(pr_df["Year"], pr_df["USA Precipitation (mm)"])
plt.plot(pr_df["Year"], pr_df["AUS Precipitation (mm)"])
plt.plot(pr_df["Year"], pr_df["Brazil Precipitation (mm)"])
plt.xticks(rotation=45)

USA_handle, = plt.plot(pr_df["Year"], pr_df["USA Precipitation (mm)"], color = "blue", label ="USA Precip")
AUS_handle, = plt.plot(pr_df["Year"], pr_df["AUS Precipitation (mm)"], color = "red", label ="AUS Precip")
BRA_handle, = plt.plot(pr_df["Year"], pr_df["Brazil Precipitation (mm)"], color = "green", label ="BRA Precip")
plt.legend(handles=[USA_handle, AUS_handle, BRA_handle], loc = "best")

plt.show()
```

```python
pr_df.plot("Year", "USA Precipitation (mm)", kind="scatter")
```

```python
x_values = pr_df["Year"]
y_values = pr_df["USA Precipitation (mm)"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(1940,48),fontsize=15,color="red")
plt.xlabel("Year")
plt.ylabel("USA Precipitation (mm)")
plt.title("USA Precipitation (mm) vs. Year")
print(f"The r-value is: {rvalue**2}")
plt.xticks(rotation=45)
plt.savefig("time_vs_precip_figures/usa_regress.png")
plt.show()
```
